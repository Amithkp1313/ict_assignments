```python
books = {}
students = {}
while True:
    print("\n1 Add Book")
    print("2 Add Student")
    print("3 Issue Book")
    print("4 Return Book")
    print("5 Exit")

    choice = input("Choice: ")

    if choice == "1":
        bid = input("Book ID: ")
        name = input("Book Name: ")
        books[bid] = name
        print("Book added")

    elif choice == "2":
        sid = input("Student ID: ")
        name = input("Student Name: ")
        students[sid] = name
        print("Student added")

    elif choice == "3":
        sid = input("Student ID: ")
        bid = input("Book ID: ")

        if sid in students and bid in books:
            print("Book issued to", students[sid])
        else:
            print("Invalid details")

    elif choice == "4":
        sid = input("Student ID: ")
        bid = input("Book ID: ")

        if sid in students and bid in books:
            print("Book returned by", students[sid])

    elif choice == "5":
        print("Exit")
        break

    else:
        print("Wrong choice")
```

```python
expenses = []

BUDGET = 5000

while True:
    print("\n--- Personal Expense Tracker ---")
    print("1. Add Expense")
    print("2. Show Total Expense")
    print("3. Category-wise Expense")
    print("4. Highest Expense Day")
    print("5. Exit")

    choice = input("Enter choice: ")

    if choice == "1":
        date = input("Enter date (DD-MM): ")
        category = input("Enter category: ")
        amount = int(input("Enter amount: "))

        expenses.append([date, category, amount])
        print("Expense Added!")

    elif choice == "2":
        total = 0
        for e in expenses:
            total = total + e[2]

        print("Total Expense:", total)

        if total > BUDGET:
            print("âš  Budget Exceeded!")

    elif choice == "3":
        categories = {}

        for e in expenses:
            if e[1] in categories:
                categories[e[1]] = categories[e[1]] + e[2]
            else:
                categories[e[1]] = e[2]

        print("Category-wise Expense:")
        for c in categories:
            print(c, ":", categories[c])

    elif choice == "4":
        day_total = {}

        for e in expenses:
            if e[0] in day_total:
                day_total[e[0]] = day_total[e[0]] + e[2]
            else:
                day_total[e[0]] = e[2]

        max_day = ""
        max_amount = 0

        for d in day_total:
            if day_total[d] > max_amount:
                max_amount = day_total[d]
                max_day = d

        if max_day != "":
            print("Highest Expense Day:", max_day, "-", max_amount)
        else:
            print("No expenses found")

    elif choice == "5":
        print("Thank you!")
        break

    else:
        print("Invalid choice")
```
